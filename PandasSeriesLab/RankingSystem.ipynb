```python
import pandas as pd

# Create Series
scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']
s = pd.Series(scores, index=players)

# 1. Rank players (higher score = better rank)
ranks = s.rank(ascending=False)
print("Player ranks:")
print(ranks)

# 2. Display top 3 players
top3 = s.sort_values(ascending=False).head(3)
print("\nTop 3 players:")
print(top3)

# 3. Find player with lowest score
lowest_player = s.idxmin()
lowest_score = s.min()
print("\nLowest scoring player:", lowest_player, "Score:", lowest_score)
```
