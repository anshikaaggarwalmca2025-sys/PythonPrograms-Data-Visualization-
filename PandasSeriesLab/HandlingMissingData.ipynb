```python
import pandas as pd

# Create Series with missing values
data = [10, None, 30, None, 50, 60, None]
s = pd.Series(data)

# 1. Count missing values before filling
missing_before = s.isnull().sum()
print("Missing values before:", missing_before)

# 2. Fill missing values with mean
mean_value = s.mean()
s_filled = s.fillna(mean_value)

# 3. Count missing values after filling
missing_after = s_filled.isnull().sum()
print("Missing values after:", missing_after)

# 4. Drop remaining null values (if any)
s_clean = s_filled.dropna()
print("\nCleaned Series:")
print(s_clean)
```
