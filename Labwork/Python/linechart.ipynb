```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
sales = [120, 150, 170, 160, 180, 200, 210, 190, 220, 230]

# Create the plot
plt.figure(figsize=(10,5))

# Plot the line with circle markers
plt.plot(months, sales, marker='o', color='blue', linewidth=2)

# Add title and labels
plt.title("Monthly Sales (in Rs)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales in Rs", fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Annotate each point with its sales value
for month, sale in zip(months, sales):
    plt.text(month, sale + 2, str(sale), ha='center', va='bottom', fontsize=10, color='darkblue')

# Show the plot
plt.show()
```

```python

# Data: Years and corresponding company profit in millions
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
profit = [20, 25, 30, 35, 40, 38, 45, 50, 55, 60]  # in millions

# Create the plot
plt.figure(figsize=(10,6))

# Plot the line with circle markers
plt.plot(years, profit, marker='o', color='green', linewidth=2)

# Add title and axis labels
plt.title("Yearly Company Profit (Million USD)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Profit (Million USD)", fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Rotate x-axis labels for clarity
plt.xticks(years, rotation=45)

# Annotate each point with its profit value
for year, prof in zip(years, profit):
    plt.text(year, prof + 0.8, str(prof), ha='center', va='bottom', fontsize=10, color='darkgreen')

# Show the plot
plt.show()

```
