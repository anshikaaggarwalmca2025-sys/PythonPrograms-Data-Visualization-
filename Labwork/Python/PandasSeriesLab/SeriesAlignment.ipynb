```python
import pandas as pd

# Create two Series
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# 1. Add both Series (alignment happens automatically)
result = s1 + s2
print("Result with NaN values:")
print(result)

# 2. Replace NaN with 0 and recompute
result_fixed = s1.add(s2, fill_value=0)
print("\nResult after replacing NaN with 0:")
print(result_fixed)
```
