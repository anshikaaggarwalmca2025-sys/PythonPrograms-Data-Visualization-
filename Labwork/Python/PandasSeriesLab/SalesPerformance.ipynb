```python
import pandas as pd

# Create Series
sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
s = pd.Series(sales, index=days)

# 1. Find best and worst sales day
best_day = s.idxmax()
worst_day = s.idxmin()
print("Best day:", best_day, "Sales:", s.max())
print("Worst day:", worst_day, "Sales:", s.min())

# 2. Calculate total and average sales
total_sales = s.sum()
average_sales = s.mean()
print("\nTotal sales:", total_sales)
print("Average sales:", average_sales)

# 3. Apply 20% bonus where sales > 400
bonus_sales = s.copy()
bonus_sales[s > 400] *= 1.20
print("\nSales after bonus:")
print(bonus_sales)

# 4. Sort sales in descending order
print("\nSales sorted (descending):")
print(s.sort_values(ascending=False))
```
