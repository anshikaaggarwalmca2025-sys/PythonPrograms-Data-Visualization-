```python
import pandas as pd

# Create Series
data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

# Perform multiple operations in a chain
result = (
    s.fillna(s.median())        # Fill missing values with median
     .loc[lambda x: x > 20]     # Filter values greater than 20
     .sort_values(ascending=False)  # Sort in descending order
     .reset_index(drop=True)    # Reset index
)

print("Final result after chained operations:")
print(result)
```
