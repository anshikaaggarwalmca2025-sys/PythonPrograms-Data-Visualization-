```python
import pandas as pd

# Create Series
salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']
s = pd.Series(salary, index=emp)

# 1. Increase salary by 10% where salary < 30000
s[s < 30000] *= 1.10

# 2. Decrease salary by 5% where salary > 45000
s[s > 45000] *= 0.95

# 3. Calculate total salary expense
total_salary = s.sum()

print("Updated salaries:")
print(s)
print("\nTotal salary expense:", total_salary)
```
