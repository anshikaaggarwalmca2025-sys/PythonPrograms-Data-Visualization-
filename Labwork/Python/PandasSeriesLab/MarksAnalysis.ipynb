```python
import pandas as pd

# Create Series
marks = [45, 67, 89, 34, 90, 55, 72]
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS', 'Biology', 'History']
s = pd.Series(marks, index=subjects)

# 1. Find subjects with marks greater than 70
print("Subjects with marks > 70:")
print(s[s > 70])

# 2. Replace marks less than 40 with "Fail"
s_updated = s.copy()  # create a copy to avoid modifying original data
s_updated[s_updated < 40] = "Fail"
print("\nMarks after replacing < 40 with 'Fail':")
print(s_updated)

# 3. Count subjects scoring above average
average_marks = s.mean()
count_above_avg = (s > average_marks).sum()
print("\nNumber of subjects above average:", count_above_avg)
```
