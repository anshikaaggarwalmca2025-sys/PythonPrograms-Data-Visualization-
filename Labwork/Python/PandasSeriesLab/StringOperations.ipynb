```python
import pandas as pd

# Create Series
names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

# 1. Convert all names to uppercase
print("Uppercase names:")
print(s.str.upper())

# 2. Find names containing letter 'a'
print("\nNames containing 'a':")
print(s[s.str.contains('a')])

# 3. Replace 'anuj' with 'Anuj Kumar'
s.replace('anuj', 'Anuj Kumar', inplace=True)
print("\nUpdated names:")
print(s)
```
