```python
import matplotlib.pyplot as plt

# Data: Courses and number of students
courses = ["Python", "Java", "C++", "Web", "AI"]
students = [40, 35, 25, 30, 20]

# Create the plot
plt.figure(figsize=(8,5))

# Plot the bar chart
plt.bar(courses, students, color='skyblue')

# Add title and axis labels
plt.title("Number of Students in Different Courses", fontsize=14)
plt.xlabel("Course", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)

# Add grid on y-axis for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with the number of students
for i, count in enumerate(students):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=10, color='darkblue')

# Show the plot
plt.show()
```

```python
# Data: Products and their sales
products = ["Laptop", "Mobile", "Tablet", "Keyboard", "Mouse"]
sales = [120, 200, 90, 150, 180]

# Create the plot
plt.figure(figsize=(8,5))

# Plot the bar chart
plt.bar(products, sales, color='green')

# Add title and axis labels
plt.title("Sales by Product", fontsize=14)
plt.xlabel("Product", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Add grid on y-axis for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with its sales value
for i, sale in enumerate(sales):
    plt.text(i, sale + 2, str(sale), ha='center', va='bottom', fontsize=10, color='darkgreen')

# Show the plot
plt.show()
```

```python
# Data: Students and their marks
students = ["A", "B", "C", "D", "E"]
marks = [78, 85, 90, 76, 88]

# Create the plot
plt.figure(figsize=(8,5))

# Plot the bar chart
plt.bar(students, marks, color='purple')

# Add title and axis labels
plt.title("Marks of Students", fontsize=14)
plt.xlabel("Student", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.ylim(0, 100)  # Assuming marks are out of 100

# Add grid on y-axis for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with the mark
for i, mark in enumerate(marks):
    plt.text(i, mark + 1, str(mark), ha='center', va='bottom', fontsize=10, color='b')

# Show the plot
plt.show()
```
